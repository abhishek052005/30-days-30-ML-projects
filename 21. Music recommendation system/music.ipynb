```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn.model_selection import train_test_split

```

```python
data = pd.read_csv('Hindi_songs.csv')
data2 = pd.read_csv('Old_songs.csv')
```

```python
df = pd.concat([data, data2], ignore_index=True)
df.sample(5)
```

```python
df['singer'] = df['singer'].str.replace('|', ' ')
df.sample(5)
```

```python
def convert_to_second(time_str):
    if pd.isna(time_str):
        return np.nan
    time_parts = time_str.split(':')
    if len(time_parts) == 2:
        minutes, seconds = time_parts
        return int(minutes) * 60 + int(seconds)
    elif len(time_parts) == 3:
        hours, minutes, seconds = time_parts
        return int(hours) * 3600 + int(minutes) * 60 + int(seconds)
    else:
        return np.nan
    
df['duration'] = df['duration'].apply(convert_to_second)
df.sample(5)
```

```python
print(df.info())
print(df.isnull().sum())

```

```python
print(df.duplicated().sum())
df.dropna(inplace=True)
print(df.info())
print(df.isnull().sum())
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

df=df.dropna()
relevant_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                     'acousticness', 'liveness', 'Valence', 'tempo']

X = df[relevant_features]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), relevant_features)
    ])


pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('knn', NearestNeighbors(n_neighbors=5, algorithm='auto', metric='euclidean'))])
   
pipeline.fit(X)
```

```python
pipeline.named_steps['knn'].kneighbors(X.iloc[0].values.reshape(1, -1))
```

```python
def recommend_songs(song_index, n_recommendations=5):
    """
    Recommend similar songs based on KNN
    
    Parameters:
    song_index: Index of the song to get recommendations for
    n_recommendations: Number of recommendations to return
    """
    # Get distances and indices of nearest neighbors
    distances, indices = pipeline.named_steps['knn'].kneighbors(
        X.iloc[song_index].values.reshape(1, -1), n_neighbors=n_recommendations+1
    )
    
    # Remove the first result (the song itself)
    recommended_indices = indices[0][1:n_recommendations+1]
    
    print(f"\n🎵 Original Song:")
    print(f"Title: {df.iloc[song_index]['song_name']}")
    print(f"Singer: {df.iloc[song_index]['singer']}")
    print(f"\n🎶 Recommended Similar Songs:")
    print("-" * 60)
    
    recommendations_df = df.iloc[recommended_indices][['song_name', 'singer', 'danceability', 'duration', 'tempo']]
    print(recommendations_df.to_string())
    
    return recommended_indices

# Test the recommendation function
recommend_songs(0, n_recommendations=5)
```
