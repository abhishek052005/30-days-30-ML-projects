# Popularity Based Recommender System

```python
ratings_with_names = ratings.merge(books, on='ISBN')
ratings_with_names.sample(5)
```

```python
num_rating_df = ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_names = ratings_with_names.dropna(subset=['Book-Rating'])
ratings_with_names.dtypes
```

```python
avg_rating_df = (
    ratings_with_names
    .groupby('Book-Title')['Book-Rating']
    .mean()
    .reset_index()
)
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df = popularity_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating','Year-Of-Publication','Publisher']]
```

```python
popular_df
```

```python
import pickle
pickle.dump(popular_df, open('popular_books.pkl', 'wb'))
```

# Collaborative Filtering


```python
x = ratings_with_names.groupby('User-ID').count()['Book-Rating'] > 200
good_users = x[x].index
```

```python
filter_rating = ratings_with_names[ratings_with_names['User-ID'].isin(good_users)]
```

```python
y = filter_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_books)]
final_rating.duplicated().sum()
```

```python
table = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
table.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(table)  #row with row comparison  coz books are in rows
similarity_scores
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(table.index==book_name)[0][0]
    distances = similarity_scores[index]
    book_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    data = []
    for i in book_list:
        item = []
        temp_df = books[books['Book-Title']==table.index[i[0]]]
        item.append(temp_df.drop_duplicates(['Book-Title'])['Book-Title'].values[0])
        item.append(temp_df.drop_duplicates(['Book-Title'])['Book-Author'].values[0])
        item.append(temp_df.drop_duplicates(['Book-Title'])['Image-URL-M'].values[0])
        data.append(item)
    return data
print(recommend("1984"))
```
