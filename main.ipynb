```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.linear_model import LinearRegression

```

```python
data = pd.read_csv('Laptop_price.csv')
data
```

```python
data.isnull().sum()
```

```python

X = data.drop('Price', axis=1)
Y = data['Price']
```

```python
# X = pd.get_dummies(X, columns=['Brand'], drop_first=True)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X['Brand'] = le.fit_transform(X['Brand'])

X
```

```python
X_train ,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1,random_state=7)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_scaled, Y_train)

# Predictions
Y_pred = model.predict(X_test_scaled)
Y_pred
```

```python

# Predictions
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

# Train performance
rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_pred))
r2_train = r2_score(Y_train, Y_train_pred)

print("Train RMSE:", rmse_train)
print("Train R²:", r2_train)

# Test performance
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_pred))
r2_test = r2_score(Y_test, Y_test_pred)

print("Test RMSE:", rmse_test)
print("Test R²:", r2_test)
```

```python
import joblib

# Save the model
joblib.dump(model, "linear_regression_model.pkl")

# Later, load the model
loaded_model = joblib.load("linear_regression_model.pkl")

# Test prediction
print("Prediction from loaded model:", loaded_model.predict(X_test_scaled[:5]))
```

```python
# Save scaler
joblib.dump(scaler, "scaler.pkl")

# Load scaler
loaded_scaler = joblib.load("scaler.pkl")

joblib.dump(le, "label_encoder.pkl")


# Example: scale new data before prediction
new_data_scaled = loaded_scaler.transform(X_test.iloc[:5])
print("Prediction:", loaded_model.predict(new_data_scaled))
```
